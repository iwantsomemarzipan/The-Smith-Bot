```python
import pandas as pd

csv_file = 'data/songs.csv'
txt_file = 'data/lyrics_only.txt'

df = pd.read_csv(csv_file)
songs = df['Lyrics'].str.cat(sep='\n')

with open(txt_file, 'w', encoding='utf-8') as txtfile:
    txtfile.write(songs)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

with open(txt_file, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Collecting tokens
def process_line(line):
    tokens = word_tokenize(line)
    tokens = [word.lower() for word in tokens if word not in string.punctuation]
    stop_words =  set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [word for word in tokens if "'" not in word and "â€¦" not in word]
    
    return tokens
```

```python
processed_tokens = [process_line(line) for line in lines]
print(processed_tokens)
```

```python
os.chdir('./word2vec_training')
```

```python
import gensim

word_model = gensim.models.Word2Vec(processed_tokens, vector_size=100, min_count=1, window=5, epochs=100)
print('Checking similar words:')

# Checking similarity for random words
for word in ['girls', 'men', 'light', 'please', 'soon', 'freedom']:
  most_similar = ', '.join('%s (%.2f)' % (similar, dist) for similar, dist in word_model.wv.most_similar(word)[:8])
  print('  %s -> %s' % (word, most_similar))
```

```python
word_model.wv.save_word2vec_format("the_smiths_w2v_model.bin", binary=True)
```
