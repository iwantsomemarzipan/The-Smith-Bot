```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.diegocaponera.com/the-smiths-complete-lyrics'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
titles = soup.find_all('h3')
lyric_groups = soup.find_all('p')

# Списки для хранения названий и текстов песен
titles_list = []
lyrics_list = []
```

```python
for title in titles:
    # Находим все теги <p>, следующие за текущим тегом <h3> до следующего <h3>
    lyrics = []
    next_element = title.find_next_sibling()
    
    while next_element and next_element.name != 'h3':
        if next_element.name == 'p':
            # Получаем текст с сохранением разрыва строк
            lyrics.append(next_element.get_text(separator='\n').strip())
        next_element = next_element.find_next_sibling()
    
    # Сохраняем текст в единую строку
    full_lyrics = '\n'.join(lyrics)
    
    titles_list.append(title.text.strip())
    lyrics_list.append(full_lyrics)
```

```python
print(len(titles_list))
print(titles_list)
print(len(lyrics_list))
print(lyrics_list)
```

```python
ids = list(range(len(titles_list)))
```

```python
data = {'ID': ids, 'Title': titles_list, 'Lyrics': lyrics_list}
df = pd.DataFrame(data)

df.set_index('ID', inplace=True)
print(df)
```
